````python
1.5 Laws of Propositional Logic [h2]
Fundamental principles that govern the logical equivalence of compound
propositions, allowing for the substitution and manipulation of expressions

1.5.1 Substitution and Logical Equivalence [h3]
Substitution allows replacing a proposition with an equivalent one in a complex expression
ex: P → Q is equivalent to ¬P ∨ Q

1.5.2 Laws of Propositional Logic
  ─> 1.1 Overview
    ├─> Essential laws for establishing logical equivalence of compound propositions
  ─> 1.2 Key Laws
    ├─> Idempotent Law
    │   └─> Simplifies expressions where a variable is combined with itself
            (P ∧ P = P, P ∨ P = P)
    ├─> Associative Law
    │   └─> Allows rearrangement of parentheses in expressions
            (P ∨ (Q ∨ R) = (P ∨ Q) ∨ R)
    ├─> Commutative Law
    │   └─> Order of variables does not affect the expression's value
            (P ∨ Q = Q ∨ P, P ∧ Q = Q ∧ P)
    ├─> Distributive Law
    │   └─> Distributes conjunctions/disjunctions over each other
            (P ∧ (Q ∨ R) = (P ∧ Q) ∨ (P ∧ R))
    ├─> Identity Law
    │   └─> Combining a variable with a true/false identity returns the original variable
            (P ∨ False = P, P ∧ True = P)
    ├─> Domination Law
    │   └─> A variable dominated by truth or falsehood determines the expression
            (P ∨ True = True, P ∧ False = False)
    ├─> Double Negation Law
    │   └─> Negating a negation returns the original variable
            (¬(¬P) = P)
    ├─> Complement Law
    │   └─> A variable and its negation cancel out to a definite value
          (P ∨ ¬P = True, P ∧ ¬P = False)
    ├─> De Morgan's Law
    │   └─> Distributes negation across conjunctions/disjunctions
           (¬(P ∨ Q) = ¬P ∧ ¬Q, ¬(P ∧ Q) = ¬P ∨ ¬Q)
    ├─> Absorption Law
    │   └─> Simplifies expressions by absorbing redundant parts
            (P ∨ (P ∧ Q) = P, P ∧ (P ∨ Q) = P)
    └─> Conditional Law
        └─> Expresses implications in terms of disjunction
             (P → Q = ¬P ∨ Q)

Exercise 1: Participation Activity: Substituting Equivalences
  ─> 1.1 Applying Logical Equivalence
    ├─> Use logical equivalence to identify matching propositions
  ─> 1.2 Examples of Equivalence
    ├─> ¬(P ∧ Q) is equivalent to ¬P ∨ ¬Q
    │   └─> Demonstrates De Morgan's Law: negating a conjunction is equivalent to the disjunction of the negations
    ├─> P → Q is equivalent to ¬P ∨ Q
    │   └─> Shows how an implication can be represented as a disjunction
  ─> 1.3 Quality Example with Detailed Explanation
    ├─> Consider the equivalence P → Q and ¬P ∨ Q
    │   ├─> Create a truth table to verify equivalence
    │   ├─> | P     | Q     | P → Q | ¬P    | ¬P ∨ Q |
    │   ├─> | True  | True  | True  | False | True   |
    │   ├─> | True  | False | False | False | False  |
    │   ├─> | False | True  | True  | True  | True   |
    │   ├─> | False | False | True  | True  | True   |
    │   └─> This table confirms that P → Q and ¬P ∨ Q are logically equivalent, as their truth values match across all possible states of P and Q

Exercise 2: De Morgan's Laws with Full Table
  ─> 1.1 Concept Overview
    ├─> Demonstrates distribution of negation in expressions
  ─> 1.2 Law Expressions
    ├─> ¬(P ∨ Q) ↔ ¬P ∧ ¬Q
    │   └─> Negating the disjunction of P and Q equals the conjunction of their negations
    └─> ¬(P ∧ Q) ↔ ¬P ∨ ¬Q
        └─> Negating the conjunction of P and Q equals the disjunction of their negations
  ─> 1.3 Truth Table for De Morgan's Laws
    ├─> Columns: P, Q, P ∨ Q, ¬(P ∨ Q), ¬P, ¬Q, ¬P ∧ ¬Q, P ∧ Q, ¬(P ∧ Q), ¬P ∨ ¬Q
    ├─> Rows represent all truth value combinations for P and Q
    │   ├─> | P     | Q     | P ∨ Q | ¬(P ∨ Q) | ¬P    | ¬Q    | ¬P ∧ ¬Q | P ∧ Q | ¬(P ∧ Q) | ¬P ∨ ¬Q |
    │   ├─> | True  | True  | True  | False    | False | False | False   | True  | False    | False   |
    │   ├─> | True  | False | True  | False    | False | True  | False   | False | True     | True    |
    │   ├─> | False | True  | True  | False    | True  | False | False   | False | True     | True    |
    │   └─> | False | False | False | True     | True  | True  | True    | False | True     | True    |
    └─> This table demonstrates that ¬(P ∨ Q) is indeed equivalent to ¬P ∧ ¬Q and ¬(P ∧ Q) is equivalent to ¬P ∨ ¬Q,
    as per De Morgan's Laws
````

````python
1.10 More Nested Quantifiers [h2]
Explores deeper aspects of nested quantifiers, emphasizing their use in expressing complex relationships and unique scenarios in logical statements.
    ∀ (Universal Quantifier): For all
    → (Conditional Operation): Implies
    ∃ (Existential Quantifier): There exists
    ¬ (Negation): Not
    Predicate: A logical statement dependent on variables
    Universal Quantifier: Asserts truth for all values in domain

1.10.1 Using Logic to Express 'Everyone Else' [h3]
Logic can express complex relationships among multiple entities in a domain
ex: "∀x ∀y (x ≠ y → P(x, y))" is true indicating everyone sent an email to everyone else

Guide to Interpreting: "Everyone Sent an Email to Everyone Else"
Logical Expression: "∀x ∀y (x ≠ y → P(x, y))"
Components:
∀x ∀y: For every person x and every person y.
x ≠ y: x and y are not the same person.
→ P(x, y): If the condition is met (x ≠ y), then person x has sent an email to person y.
Plain English: Every person sent an email to every other person, excluding themselves.
How to Verify: To verify this using a cross-tab table, check that all cells
 (excluding those where a person would send an email to themselves) are marked true.

Cross-Tab Table Interpretation Example: [h4]
Note: '-' indicates a person does not send an email to themselves.
+-------+-----+-----+-----+
|       |  A  |  B  |  C  |
+-------+-----+-----+-----+
|   A   | -   |true |true |
+-------+-----+-----+-----+
|   B   |true | -   |true |
+-------+-----+-----+-----+
|   C   |true |true | -   |
+-------+-----+-----+-----+

1.10.2 Expressing Uniqueness [h3]
Logic to specify singular occurrences within a group
ex: "∃x (L(x) ∧ ∀y (x ≠ y → ¬L(y)))" indicates exactly one person was late

Mini Guide to Interpreting "Expressing Uniqueness"

Understanding the Statement: [h4]
- "∃x (L(x) ∧ ∀y (x ≠ y → ¬L(y)))" is a logical expression used to denote that there is exactly one individual within a group who possesses a unique attribute, in this case, being late.

Breaking Down the Components: [h4]
- "∃x": This signifies that there exists at least one x in the domain.
- "L(x)": L(x) indicates that the individual x was late.
- "∀y": For all other individuals y in the domain.
- "x ≠ y": x is not the same individual as y.
- "¬L(y)": None of the other individuals y were late.

Plain English Interpretation: [h4]
- The statement translates to "There exists one and only one person who was late, and all others were on time."
- This logically asserts the uniqueness of the occurrence, meaning exactly one person in the group has the attribute of being late, distinguishing them from the rest.

Example Scenario: [h4]
- In a group of three people (A, B, C), if A was late and both B and C were on time, the condition is satisfied. The table or truth table to visualize this scenario would help confirm that only one person meets the criteria set out by the logical expression.

Application: [h4]
- This concept is crucial when needing to prove or assert the existence of a unique element within a set, based on a specific condition or attribute.
- Useful in mathematical proofs, database queries, or any scenario where identifying an exclusive occurrence within a group is necessary.

Note: Interpreting and applying this logical statement requires careful consideration of the domain and the conditions being examined to ensure accurate conclusions.


1.10.3 Moving Quantifiers [h3]
Shifting quantifiers in logical expressions and implications
ex: Moving quantifiers can change meaning: "∀x ∃y M(x, y)" is not equivalent to "∃y ∀x M(x, y)"

Mini Guide to Interpreting: Moving Quantifiers

Understanding Quantifier Movement: [h4]
   ─> Quantifiers like ∀ (forall) and ∃ (exists) specify the scope of variables in logical statements.
   ─> Changing the order of these quantifiers alters the statement's meaning and implications.

Example with Explanation: [h4]
   ├─> Original Statement: "∀x ∃y M(x, y)"
   │   └─> Meaning: For every x, there exists a y such that M(x, y) holds true. This implies a dependency of y on x.
   │       └─> Plain English: "For each person x, there is at least one task y they can complete."

   ├─> Modified Statement: "∃y ∀x M(x, y)"
   │   └─> Meaning: There exists a y for which, for every x, M(x, y) holds true. This suggests y is independent of x.
   │       └─> Plain English: "There is at least one task y that every person x can complete."

Key Insight: [h4]
   ─> The order of quantifiers critically affects the statement's scope and what it asserts about the relationship between variables.
   ─> Moving from "∀x ∃y" to "∃y ∀x" shifts from an individual-specific context to a universal one, changing the nature of the relationship between x and y.



1.10.4 Nested Quantifiers in English [h3]
Translating logical expressions into comprehensible language
ex: "∀x ∃y K(x, y)" means every student knows someone's phone number

Mini Guide to Interpreting: Nested Quantifiers in English

Understanding Nested Quantifiers: [h4]
   ─> Nested quantifiers allow for the expression of complex relationships by combining universal (∀) and existential (∃) quantifiers.

Example with Explanation: [h4]
   ├─> Statement: "∀x ∃y K(x, y)"
   │   └─> Meaning: For every x, there exists at least one y such that the predicate K(x, y) is true.
   │       └─> Plain English: "Every student (x) knows at least one person's (y) phone number."

Key Insight: [h4]
   ─> This logical structure is used to convey that within a group (all students), each member has a specific connection or knowledge related to another group member (someone's phone number), highlighting the capability of nested quantifiers to detail individual-specific yet universally applicable relationships.


1.10.5 Cross-Tab Tables [h3]
Tables illustrating relationships among entities in nested quantifiers. Used to understand complex logical relations
ex: if 'A' sent an email to 'B'(P(A, B) = true) and not to 'C'(P(A, C) = false), the table helps visualize these relations

Mini Guide to Interpreting: Cross-Tab Tables for Nested Quantifiers

Purpose of Cross-Tab Tables: [h4]
   ─> Cross-tabulation tables (cross-tab tables) visually represent the relationships between entities, especially useful in logical expressions involving nested quantifiers.

Constructing a Cross-Tab Table: [h4]
   ─> Rows and columns represent entities (e.g., people, items). Each cell shows the relationship (true/false) between the row entity and column entity.

Example Table: Email Sent Scenario: [h4]
   +-------+-----+-----+-----+
   |       |  A  |  B  |  C  |
   +-------+-----+-----+-----+
   |   A   | true|true |false|
   +-------+-----+-----+-----+
   |   B   |false|true |true |
   +-------+-----+-----+-----+
   |   C   |true |false|true |
   +-------+-----+-----+-----+

   ├─> Interpretation:
   │   ├─> Row 'A' shows A sent an email to themselves (A) and to B, but not to C.
   │   ├─> Row 'B' shows B did not send an email to A, but sent to themselves (B) and to C.
   │   └─> Row 'C' shows C sent an email to A and themselves (C), but not to B.

Key Insight: [h4]
   ─> Cross-tab tables offer a clear visual method to understand complex logical relationships, making it easier to grasp who has interacted with whom, based on specified criteria like sending emails. This tool is invaluable for analyzing relations defined by logical statements, particularly those involving nested quantifiers.


Exercise 1: Assessing Cross-Tab Tables [h3]
Use tables to evaluate logical statements with nested quantifiers
1) "Everyone sent an email to someone else" (True/False based on table data)
2) "∀x ∃y P(x, y)" (True/False based on table data)
3) If x = A, who will y be for "∀y P(x, y)"? (Answer depends on table data)
4) "∀x ∀y (x ≠ y → P(x, y))" (True/False)

|     | A     | B     | C     |
|-----|-------|-------|-------|
| A   | False | True  | True  |
| B   | True  | False | True  |
| C   | True  | True  | False |


Exercise 2: Expressing Uniqueness with Cross-Tab Tables [h3]
Tables to show unique occurrences in a domain
ex: "∃x (L(x) ∧ ∀y (x ≠ y → ¬L(y)))" can be visualized in a table to determine if exactly one person was late

|     | Late   |
|-----|--------|
| A   | True   |
| B   | False  |
| C   | False  |


Exercise 3: Uniqueness in Cross-Tab Tables [h3]
Use tables to determine uniqueness in scenarios
1) "Exactly one person was late for the meeting" (True/False based on table data)
2) "∃x (L(x) ∧ ∀y (x ≠ y → ¬L(y)))" (True/False based on table data)

1) True - Exactly one person was late for the meeting (A was late, B and C were not).
2) True - There exists one person (A) who was late, and for all others (B, C), they were not late, fitting the logical expression '∃x (L(x) ∧ ∀y (x ≠ y → ¬L(y)))'.



1.10.5.1 Cross-Tab Table Examples [h3]
1. Email Sent Scenario [h4]
Imagine a small group of three people (A, B, C) and a table indicating whether each person has sent an email to another.

Plain English: Person A sent emails to A (themselves) and B but not to C. Person B didn't send an email to A but did to B and C, and so on.

|       |  A    |  B    |  C    |
|-------|-------|-------|-------|
|   A   | true  | true  | false |
|   B   | false | true  | true  |
|   C   | true  | false | true  |

This table helps students visualize the concept of direct and indirect relationships within a group, relevant to understanding nested quantifiers in logical expressions.

2. Attendance Scenario [h4]
A group of three people (A, B, C) and a table indicating their attendance at a meeting.

Plain English: A was present when A and C were there but absent when B was there, and so on.

|       |   A    |   B    |   C    |
|-------|--------|--------|--------|
|   A   | present| absent |present |
|   B   | absent |present |absent  |
|   C   | present| absent |present |

This scenario demonstrates the presence and absence of individuals in different settings, aiding in the comprehension of universal and existential quantifiers through real-world applications.

3. Project Collaboration Scenario [h4]
Three team members (A, B, C) working on a project, with a table showing who has collaborated with whom.

|       |  A    |  B    |  C    |
|-------|-------|-------|-------|
|   A   | true  | false | true  |
|   B   | true  | true  | false |
|   C   | false | true  | true  |

Plain English: Each row and column intersection shows whether the person on the left has collaborated with the person on top. For example, A has collaborated with C but not with B.

These tables not only visually represent complex logical relationships but also ground abstract logical concepts in everyday scenarios, making them more accessible to students learning about nested quantifiers.
````

````python
1.12 Rules of Inference with Propositions [h2]
   ─> Rules of inference provide a framework for deducing conclusions from given premises
   ─> Argument: A sequence of propositions (hypotheses) followed by a conclusion
   ─> Validity: An argument is valid if the conclusion is  true when all hypotheses are true
   ─> Truth Table: A table showing the truth value of   propositions under different hypotheses
   ─> Modus Ponens: A rule of inference where if "P → Q" and "P" are both true, then "Q" is true

1.12.1 Rules of Inference [h3]
Rules of inference are pre-established  valid arguments used in logical proofs
ex: Modus ponens: "If P, then Q. P is true. Therefore, Q is true." is a valid rule of inference

Example: Modus Ponens [h4]
   ├─> Rule: If "P → Q" (If P, then Q) and "P" are both true, then "Q" must also be true
   │   └─> Scenario: "If it rains, then the ground will be wet. It rains."
   │       └─> Conclusion: "Therefore, the ground will be wet."
   └─> Interpretation: This rule allows us to infer the truth of "Q" (the ground will be wet) based on the truth of "P" (it rains) and the conditional statement "P → Q" (If it rains, then the ground will be wet).

Key Insight: [h4]
   ─> Modus Ponens is one of many rules of inference that validate arguments in logical proofs, ensuring that conclusions logically follow from premises.

1.12.2 Proving Argument Validity [h3]
To validate arguments, we utilize rules of inference

Example: [h4]
Consider the argument "If it rains, then the ground is wet. It is raining. Therefore, the ground is wet."

Steps to prove validity:
1. "If it rains (P), then the ground is wet (Q)" - Establishes a conditional relationship between rain and the ground's wetness.
2. "It is raining (P)" - Affirms the condition that it is raining.
3. Therefore, "The ground is wet (Q)" - Applying Modus Ponens, since the conditional statement and its condition are true, the conclusion must also be true.

1.12.3 English Arguments and Inference Rules [h3]
Apply inference rules to arguments phrased in English

Example: [h4]
The English argument "If it rains, I carry an umbrella. It is raining. Therefore, I am carrying an umbrella." validates using Modus Ponens

Exercise 1: Logical Proof Example [h3]
Demonstrate applying rules of inference in logical proofs

Proof example for "If P, then Q. P. Therefore, Q":
1) P (Hypothesis) - Acknowledges the initial condition
2) If P, then Q (Hypothesis) - Establishes a conditional link between P and Q
3) Q (Conclusion) - By Modus Ponens, from 1 and 2, concludes Q is true

Exercise 2: Proof Justification [h3]
Justify each step in a logical proof with the appropriate rules of inference

Exercise 3: English Argument Proof [h3]
Demonstrate the validity of English arguments by converting them into logical forms and applying rules of inference

Argument: "If it rains, I take an umbrella. It is raining. Therefore, I am taking an umbrella."

Logical translation: [h4]
1. If it rains (P), then I take an umbrella (Q)
2. It is raining (P)
3. Therefore, I am taking an umbrella (Q)

Justification: [h4]
- Steps 1 and 2 are translated into logical propositions
- Applying Modus Ponens to these propositions justifies the conclusion (Step 3), affirming the argument's validity

````
